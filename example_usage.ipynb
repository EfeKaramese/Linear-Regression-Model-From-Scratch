```python
import numpy as np
from linear_regression import CustomLinearRegression
```

```python
from sklearn.linear_model import LinearRegression

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]]) 

y = np.dot(X, np.array([1, 2])) + 3

model2 = LinearRegression()
model2.fit(X,y)
model2.predict(np.array([[3, 5]]))
```

```python
model1 = CustomLinearRegression()
model1.fit(X,y)
model1.predict(np.array([[3, 5]]))
```

```python

import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# 1. DATA GENERATION
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# 2. TRAINING THE MODEL
model = CustomLinearRegression()
model.fit(X, y)

# Create smooth line for plotting
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_line = model.predict(X_line)

# 3. VISUALIZATION 
plt.figure(figsize=(14, 5))

# --- Plot 1: Loss History ---
plt.subplot(1, 2, 1)
plt.plot(model.loss_history, color='blue', linewidth=2)
plt.title('Model Learning Process (Loss History)')
plt.xlabel('teration Number')
plt.ylabel('Hata (MSE)')
plt.grid(True, alpha=0.3)

# --- Plot 2: Regression Fit ---
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', alpha=0.5, label='Real Data')
plt.plot(X_line, y_line, color='red', linewidth=3, label='My Model')
plt.title('Linear Regression Fit')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.savefig('images/comparison_plot.png', dpi=300)
plt.show()

# Print Final Results
print(f"Final Loss (MSE): {model.loss_history[-1]:.4f}")
print(f"Learned Weight(W): {model.weights[0][0]:.4f}")
print(f"Learned Bias(b): {model.bias:.4f}")
```
